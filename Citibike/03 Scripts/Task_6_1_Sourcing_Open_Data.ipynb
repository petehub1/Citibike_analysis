# 03 Data Wrangling and cleaning

```python
citi_b.head()
```

```python
#checking null values
citi_b.isnull().sum()
```

```python
# Create a subset where the 'year of birth' column contains NaN values

birth_nan = citi_b[citi_b['birth_year'].isnull() ==True]
```

```python
birth_nan.shape
```

```python
citi_new = citi_b[citi_b['birth_year'].isnull() ==False]
```

```python
citi_new.shape
```

```python
citi_new.head()
```

```python
# drop bike_id column
citi_new = citi_new.drop(columns=['bike_id'])
```

```python
citi_new.head()
```

```python
# drop trip_id column
citi_new = citi_new.drop(columns=['trip_id'])
```

```python
citi_new.head()
```

```python
print(citi_new)
```

```python
#get info 
citi_new.info()
```

```python
#changing gender type to string
citi_new['gender'] = citi_new['gender'].astype('str')
```

```python
citi_new.info()
```

```python
# Check the unique values in the 'gender' column
print(citi_new['gender'].unique())
```

```python
# Define a mapping dictionary for gender values
gender_mapping = {'1': 'Male', '2': 'Female', '0': 'unknown'}
```

```python
# Replace gender values with their corresponding names
citi_new['gender'] = citi_new['gender'].replace(gender_mapping)
```

```python
# Verify the changes
print(citi_new['gender'].unique())
```

```python
citi_new.head()
```

```python
# creating a subset of users born after 1922
citi_new_by = citi_new.loc[citi_new['birth_year'] > 1922]
```

```python
citi_new_by.shape
```

```python
# Convert 'start_time' to datetime and rename the column to 'start_of_ride' using .loc to avoid SettingWithCopyWarning
citi_new_by.loc[:, 'start_of_ride'] = pd.to_datetime(citi_new_by['start_time'])


```

```python
citi_new.head()
```

```python
# changing end time to end of ride and making the datatype - datetime
citi_new_by['end_of_ride']=pd.to_datetime(citi_new_by['end_time'],infer_datetime_format=True) 
```

```python
# Convert 'end_time' to datetime and rename the column to 'end_of_ride' using .loc to avoid SettingWithCopyWarning
citi_new_by.loc[:, 'end_of_ride'] = pd.to_datetime(citi_new_by['end_time'])
```

```python
citi_new_by.head()
```

```python
# drop start_time column
citi_new_by = citi_new_by.drop(columns=['start_time'])
```

```python
# drop end_time column
citi_new_by = citi_new_by.drop(columns=['end_time'])
```
