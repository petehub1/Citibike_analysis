# 03 Create a correlation matrix heatmap (colored).
Discuss what the coefficients in the plot mean in terms of the relationships between the variables.

```python
# Create a subset
df_selected.columns
```

```python
sub = df_selected[['start_hour', 'start_station_latitude', 'start_station_longitude', 'end_station_latitude','end_station_longitude', 'trip_duration', 'birth_year']]
```

```python
sub
```

```python
print(sub.dtypes)
```

```python
# Calculate the correlation matrix
correlation_matrix = sub.corr()
```

```python
correlation_matrix
```

```python
# Create a heatmap
plt.figure(figsize=(12, 8))
heatmap = sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', linewidths=0.5)

# Add annotations manually
for i in range(correlation_matrix.shape[0]):
    for j in range(correlation_matrix.shape[1]):
        plt.text(j + 0.5, i + 0.5, f"{correlation_matrix.iloc[i, j]:.2f}", 
                 ha='center', va='center', color='black', fontsize=10)

plt.title('Correlation Matrix Heatmap')

# Export and save the heatmap as an image
heatmap.get_figure().savefig("heatmap.png", dpi=300)

# Display the heatmap
plt.show()

```

# The strongest correlations are observed between spatial coordinates (latitude and longitude) of the start and end stations. For instance, start_station_latitude and end_station_latitude have a moderate positive correlation (0.58).
Temporal variables like start_hour and demographic variables like birth_year show very weak or negligible correlations with other variables.
trip_duration does not show any significant correlation with any other variable in the dataset.

# Which stations are the most popular for renting bikes? Which stations are the top destinations for bike returns?


```python
# Create bar plots for the top 10 starting and ending stations

start_station_counts = df_selected['start_station_name'].value_counts().head(10)
end_station_counts = df_selected['end_station_name'].value_counts().head(10)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
start_station_counts.plot(kind='bar', color='coral')
plt.xlabel('Starting Station')
plt.ylabel('Number of Rides')
plt.title('Top 10 Most Popular Starting Stations')

plt.subplot(1, 2, 2)
end_station_counts.plot(kind='bar', color='teal')
plt.xlabel('Ending Station')
plt.ylabel('Number of Rides')
plt.title('Top 10 Most Popular Ending Stations')

plt.tight_layout()

# Save the subplot as a PNG file
plt.savefig('top_stations.png', dpi=300)

# Display the plots
plt.show()
```

# What are the peak times of day for bike rentals?

```python
# Create a bar plot (count plot) of start_hour vs. weekday

# Create the countplot
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='start_hour', hue='weekday', palette='Dark2')
plt.xlabel('Start Hour')
plt.ylabel('Count')
plt.title('Distribution of Start Hour by Weekday')
plt.legend(title='Days of the Week', loc='upper right')

# Save the figure to a file 
plt.savefig('countplot_start_hour_by_weekday.png')

# Display the plot if needed
plt.show()
```

# The peak time is 8-9 AM, and 4-7 PM.
